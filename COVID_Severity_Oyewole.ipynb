```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_df = pd.read_csv('C:/Users/User/Desktop/FY-Project/526 mat/COVID_Data.csv')
covid_df.head()
#covid_df.shape
```

```python
covid_dft_df = covid_df.drop(covid_df.iloc[:, 8:37], axis=1)
covid_dft_df
```

```python
covid = covid_dft_df.drop(['Death Interval'], axis=1)
covid.head()
```

```python
covid= covid.rename(columns={'Outcome':'Class'})
```

```python
X = covid.iloc[:, [0,1,2,3,4,5,6,7]]
y = covid.iloc[:, 8]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
print("training data: ", X_train.shape)
print("test data: ", X_test.shape)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn import svm
from sklearn.svm import SVC

clf = svm.SVC(kernel='linear', gamma='auto', C=2)
clf.fit(X_train, y_train)

y_predict = clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('Accuracy score with linear kernel:')
print(metrics.accuracy_score(y_test, y_predict))
```

```python
clf = svm.SVC(kernel='poly', degree=4)
clf.fit(X_train, y_train)

y_predict = clf.predict(X_test)

print('Accuracy score with poly kernel:')
print(metrics.accuracy_score(y_test, y_predict))
```

```python
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train)

y_predict = clf.predict(X_test)

print('Accuracy score with rbf kernel:')
print(metrics.accuracy_score(y_test, y_predict))
```
